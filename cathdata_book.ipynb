```python
#!/usr/bin/env python

# test python script
import pandas as pd

csv1_filename = "ec_terms_in_sfam.csv"

uniq_ec_counts = {}

with open(csv1_filename) as csv1_fh:
    for line in csv1_fh:
#         print(line)
        # split by special characters
        line = line.replace('"', '')
        line = line.strip()
        uniprot_acc, ec_code = line.split(',')
        # skip header line
        if uniprot_acc == "UNIPROT_ACC":
            continue
        # check output
#         print(uniprot_acc, ec_code)
        # count instances of each EC code
        uniq_ec_counts[ec_code] = uniq_ec_counts.get(ec_code, 0) + 1
        
# print out to check structure of data
uniq_ec_counts

# display data as a list ordered by EC number
# uniq_ec_items = list(uniq_ec_counts.items())
# uniq_ec_items.sort()
# print(uniq_ec_items)

# sort dict by value
for ec_code, ec_count in sorted(uniq_ec_counts.items(), key=lambda item: item[1]):
    print("%s: %s" % (ec_code, ec_count))
```
