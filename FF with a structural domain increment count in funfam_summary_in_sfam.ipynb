```python
##AIM 1. Find a funfam with a structural domain, increment count



# !/usr/bin/env python

# test python script

## Set() creats an empty set, so it should come first. Then, the counts starts from 0.




import pandas as pd
csv4_filename = "funfam_summary_in_sfam.csv"
with open (csv4_filename) as csv4_fh:
    funfam_number = set()
    funfam_count = 0
    for line in csv4_fh:
        line = line.replace('"', '')
        line = line.strip()
        FUNFAM_NUMBER,REP_ID,REP_SOURCE_ID,NUM_MEMBERS_IN_FUNFAM = line.split(',')
#         print("vars:", FUNFAM_NUMBER,REP_ID,REP_SOURCE_ID,NUM_MEMBERS_IN_FUNFAM)


## if rep source id is cath
    
        if REP_SOURCE_ID == "cath":
            print("'cath' match", line)
            funfam_number.add(FUNFAM_NUMBER)
            funfam_count += 1
        else:
            print("line does not contain 'cath'", line)
print("funfam_count (incremental count)", funfam_count)                  
print("unique funfam numbers (using set)", len(funfam_number))
print("set contents:", funfam_number)
```
