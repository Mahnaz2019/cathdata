```python
##AIM 1: skip header line
##AIM 2: count instances of each GO_ACC code
##AIM 3: filter unique GO numbers
##AIM 4: display data as a list ordered by go_acc number
##AIM 5: sort dict by value

#!/usr/bin/env python

# test python script
import pandas as pd

csv2_filename = "go_terms_in_sfam.csv"

#in this command, I will give a name to the set {} that I am going to create. Here, I want to see only the unique GO_ACC numbers:
uniq_go_acc_counts = {}



with open(csv2_filename) as csv2_fh:
    for line in csv2_fh:
        
        ## split by special characters
        line = line.replace('"', '')
        line = line.strip()
#         print(line)


##Here, I should define what each line is. This table has only one column so all the lines are GO_ACC.
        go_acc = line  
    

 
#print number of uniq_ec
print("number of go_acc strings: ", len(line))
        
        ## skip header line.  So, go_acc is defined in the previous line what it is. 
        if go_acc == "GO_ACC":
            continue
   
## check output
        print(go_acc)
        ##count instances of each GO_ACC code


      ##indention is important in Python. So, if a command is not a continution of previous
##order, the upcoming commands should be outside the line/tab.
        uniq_go_acc_counts[go_acc] = uniq_go_acc_counts.get(go_acc, 0) + 1
        
## print out to check structure of data
uniq_go_acc_counts


## display data as a list ordered by go_acc number
uniq_go_acc_items = list(uniq_go_acc_counts.items())
uniq_go_acc_items.sort()
print(uniq_go_acc_items)


## sort dict by value
for go_acc, go_count in sorted(uniq_go_acc_counts.items(), key=lambda item: item[1]):
    print("%s: %s" % (go_acc, go_count))

    
    
    ##Eventhough, there is only one column exits but we have to put 2x %S: one for go, and one for the numbers.
```
